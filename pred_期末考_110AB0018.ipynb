#(1).請問3種機器學習演算法【KNN，決策樹，隨機森林法】，用測試數據計算準確率最高的，是哪一種演算法？準確率多少？

```python
import pandas as pd
df=pd.read_csv("https://acupun.site/lecture/predict/example/resource/wine-chi-1.csv")
df
```

```python
import sklearn.model_selection as msel
x_train,x_test,y_train,y_test= msel.train_test_split(
    df[["酒精濃度","蘋果酸","灰","灰的鹼度","鎂","總酚","黃酮類化合物","非黃烷類酚類","原花青素","色彩強度","色調","稀釋酒","脯氨酸"]],
    df[["種類"]],
    test_size=0.3
    )
#x_test
```

```python
import sklearn.tree as tree
model=tree.DecisionTreeClassifier()
model.fit(x_train,y_train)
#y_pred = model.predict(x_test)
```

```python
print("決策樹測試數據的準確率(by LabelEncoder):",model.score(x_test, y_test))
```

```python
import sklearn.ensemble as es
model=es.RandomForestClassifier()
model.fit(x_train,y_train)
#y_pred = model.predict(x_test)
```

#(3).評估AI模型的成效：畫出【混淆矩陣圖】

```python
import sklearn.model_selection as msel
x_train, x_test, y_train, y_test = msel.train_test_split(
    df2[["年齡","性別2","血壓2","膽固醇2","血液鈉鉀比例"]],
    df2["用藥2"],
    test_size=0.3
   )
```

```python
#KNN模型
```

```python
#1.步驟1：先建立數學模型(預設鄰居為5個)
import sklearn.neighbors as nb
model=nb.KNeighborsClassifier(n_neighbors=5)
#2.步驟2：讓模型model學習歷史數據
model.fit(x_train,y_train)
#3.步驟3：做預測
y_pred = model.predict(x_test)
```

```python
#定量評估model的好壞(3種定量方法):整體準確率accuracy，混亂矩陣，綜合分類分析報告
#方法1:整體準確率：accuracy
##指令1:整體準確率：model.score()
##指令2:整體準確率：accuracy_score()

#方法2:混亂矩陣
##指令1:混亂矩陣：confusion_matrix()
##指令2:混亂矩陣：ConfusionMatrixDisplay()

#方法3:綜合分類分析報告
##指令1:綜合分類分析報告：classification_report()
```

```python
#辨別混淆矩陣:
##0.常見的二元分類（Binary Classification）可以使用下列四種狀態來描述模型的預測：
#TP (True Positive)：真實標籤是「正」，模型也預測為「正」。
#TN (True Negative)：真實標籤是「負」，模型也預測為「負」。
#FP (False Positive)：真實標籤是「負」，但模型誤預測為「正」（「誤報」）。
#FN (False Negative)：真實標籤是「正」，但模型誤預測為「負」（「漏報」）。

##1.整體準確度：透過對角線總和 / 全部樣本

##2.召回率 (Recall):Row=真實標籤=在所有"真實的正樣本"中，模型能夠正確找出的比例
##解讀:如果召回率高，代表模型很少漏掉真實為正的樣本（FN 少）。舉例來說，在醫學檢測中，如果我們希望盡可能偵測到「有病」的病患，就會要求模型的召回率要高，避免「漏診」。

##3.精確率 (Precision):Column=預測標籤=在"模型預測為正的樣本"中，真正正確的比例
##解讀:如果準確率（Precision）高，代表模型在判斷「是正」時很少「誤報」(FP 少)。比如垃圾郵件過濾器若要保證誤判「正常郵件」為「垃圾信」的情況很少，就需要高 Precision。

#Accuracy = (TP + TN) / (TP + TN + FP + FN)--平均而言中靶心
#Precision = TP / (TP + FP)--精準不亂飄
#Recall = TP / (TP + FN)
```

```python
#這邊用方法2:混亂矩陣
import sklearn.metrics as mts
mts.ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
#Step1:分割資料
import sklearn.model_selection as ms
x_train,x_test,y_train,y_test= ms.train_test_split(df4[["酒精濃度","蘋果酸","灰","灰的鹼度","鎂","總酚","黃酮類化合物","非黃烷類酚類","原花青素","色彩強度","色調","稀釋酒","脯氨酸"]],
                          arr4,
                          test_size=0.3
                          #train_test=0.7
                          )
```

```python
#Step2:建立模型，訓練數據
import tensorflow as tf
model=tf.keras.Sequential([
    tf.keras.layers.Dense(units=10,input_dim=13,activation='relu'),
    tf.keras.layers.Dense(units=10,activation='relu'),
    tf.keras.layers.Dense(units=3,activation='softmax')
])
```

```python
#Step3:設定類神經網絡模型的3個數學函數
model.compile(
    loss=tf.keras.losses.CategoricalCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(),
    metrics=tf.keras.metrics.CategoricalAccuracy()
)
```

```python
#Step4:讓模型model學習自變數
his_train=model.fit(x_train,y_train,epochs=1000)
```
